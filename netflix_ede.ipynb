```python
import pandas as pd
import plotly.express as px
```

```python
data_import = pd.read_csv('netflix_titles.csv')
data_import["date_added"] = pd.to_datetime(data_import["date_added"].str.strip(), format="%B %d, %Y")
data_import
```

```python
data_import.dtypes
```

## date Analysis

```python
data_import["release_year"].hist() #matplotlib version
```

```python
px.histogram(data_import, x="release_year") #plotly version
```

```python
data_import[data_import["release_year"] == 1925]
```

```python
data_import["date_added"].hist() #matplotlib version
```

```python
# extracting month from date added
data_import["date_added_month"] = data_import["date_added"].dt.month.fillna(0)
data_import["date_added_day"] = data_import["date_added"].dt.day.fillna(0)
```

```python
px.histogram(data_import, x="date_added", color = "date_added_month") #plotly version
```

```python
px.histogram(data_import, x = "date_added_month", color="type")
```

```python
px.histogram(data_import, x = "date_added_day", color="type")
# looks like 1st day of the month is when they added most
# looks like movies are much more than tv shows
```

# string column analysis

```python
data_import.columns
```

```python
data_import["country"].str.split(",",expand=True)
```

```python
country_count = data_import.copy()
country_count = pd.concat([country_count, data_import["country"].str.split(",",expand=True)], axis=1)
country_count = country_count.melt(id_vars=["type","title"], value_vars=range(12), value_name="Country")
country_count = country_count[country_count["Country"].notna()]
country_count["Country"] = country_count["Country"].str.strip()
country_count
```

```python
px.histogram(country_count, "Country", color= "type").update_xaxes(categoryorder="total descending")
# uk,japan & south korea produce more tv shows then india
```

```python
px.histogram(data_import, "rating")
```

```python
data_import["cast count"] = data_import["cast"].str.split(",")
data_import = data_import[data_import["cast count"].notna()]
data_import["cast count"] = data_import["cast count"].apply(lambda x: len(x))
data_import
```

```python
px.histogram(data_import,"cast count").update_xaxes(categoryorder = "total descending")
# in most movies or tv shows mostly 10 cost members are given credit for
```

```python
data_import["listed_in"].unique()
```

```python
data_import["listed_in"].str.split(",",expand=True)
```

```python
list_count = data_import.copy()
list_count = pd.concat([list_count, data_import["listed_in"].str.split(",",expand=True)], axis=1)
list_count = list_count.melt(id_vars=["type","title"], value_vars=range(3), value_name="list count")
list_count = list_count[list_count["list count"].notna()]
list_count["list count"] = list_count["list count"].str.strip()
list_count
```

```python
px.histogram(list_count, "list count", color= "type").update_xaxes(categoryorder="total descending")
#count of movies or tv shows besed on the categories
```
